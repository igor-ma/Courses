```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

DATA_PATH_1 = "../data/chocolura.csv"
DATA_PATH_2 = "../data/alucel.csv"
```

## Analyzing Chocolura

```python
df = pd.read_csv(DATA_PATH_1)
df.rename(columns={'vendas': 'sales'}, inplace=True)
```

```python
dec = seasonal_decompose(df['sales'], period=3)
ax = dec.plot()
```

## Analyzing Alucel

```python
df = pd.read_csv(DATA_PATH_2)
df.rename(columns={'vendas': 'sales'}, inplace=True)
df
```

```python
df['moving_average'] = df['sales'].rolling(window=3).mean()
df.plot()
```

```python
df['moving_average'] = df['sales'].rolling(window=14).mean()
df.plot()
```

```python
dec = seasonal_decompose(df['sales'], period=3)
ax = dec.plot()
```

```python
dec = seasonal_decompose(df['moving_average'].fillna(df['moving_average'].median()), period=1)
ax = dec.plot()
```
